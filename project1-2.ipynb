### 2.1 기초 분석

```python
# csv 파일 불러오기
import pandas as pd

d_df = pd.read_csv('C:/Users/hmins/Desktop/project_df.csv',encoding='cp949')
d_df[0:5]
```

```python
# 소규모 토픽 제거
d_df = d_df.loc[(d_df.topic == '정치')|
              (d_df.topic == '경제')|
              (d_df.topic == '국제')|
              (d_df.topic == '사회')|
              (d_df.topic == '문화')|
              (d_df.topic == '연예')|
              (d_df.topic == '스포츠')]

len(d_df)
```

```python
from konlpy.tag import Okt

twt = Okt()
```

```python
# 동아일보 제목에 나온 단어

d_all_word_a = []
for n in range(len(d_df)):
    text = d_df['title'].iloc[n]
    words = twt.pos(text)
    for i, j in words:
        if j == 'Noun':
            
            d_all_word_a.append(i)

d_all_word_a[1:50]
```

```python
# 최근 기사 1000개 최대 개수 단어 top 10

d_all_word_df = pd.DataFrame(d_all_word_a, columns = ['words'])
d_all_word_df = pd.DataFrame(d_all_word_df.groupby('words')['words'].count())

d_all_word_df.rename(columns= {'words':'count'}, inplace=True)
d_all_word_df.sort_values(by='count',ascending=False)[:50]
```

```python
# 동아일보 불용어 제거 top 10

d_all_word_k = []
stop_words = ['명','것','전','세','도','첫','등','안','중','개월']

for n in range(len(d_df)):
    text = d_df['title'].iloc[n]
    words = twt.pos(text)
    
    for i, j in words:
        if j == 'Noun':
            if i not in stop_words:
                d_all_word_k.append(i)
                
d_all_word_df = pd.DataFrame(d_all_word_k, columns = ['words'])
d_all_word_df = pd.DataFrame(d_all_word_df.groupby('words')['words'].count())

d_all_word_df.rename(columns= {'words':'count'}, inplace=True)
d_all_word_df.sort_values(by='count',ascending=False)[:60]
```

### 2-2. 교차표 생성

```python
d_all_word = pd.DataFrame()
stop_words = ['것','이','전','김','등','중','위','개','사','프로필','가렵니','세','단'
             ,'명','것','또','재','태','김','가람','어디','수','내','파','첫','기요',
             '사키','해','중','이','번','곳','중','새','체','위키','때','왜','서나',
             '밍','집','강','구','날','박','판','린다','무','옴','땐','안','탓','트',
             '호','를','로','온','척','창','그','자체','타','최','듯','너머','놈','던',
             '뿐','슐','더','덩이','치','윤','다가','활','점','값','및','헛','뭐','보고',
             '겉모습','유정','원','차','비','내일','깃','꿈','리','우리','공','멸','씨','진','며','바','실','규','의']

for n in range(len(d_df['title'])):
    text = d_df['title'].iloc[n]
    words = twt.pos(text)
    
    d_words_df = pd.DataFrame()
    
    for i, j in words:
        
        if j == 'Noun':
            d_words_df[i] = [1]
        
    d_all_word = pd.concat([d_all_word, d_words_df], ignore_index =True)

d_all_word[:50]
```
